```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
data = iris.data
columns = iris.feature_names
iris.head()
```

```python
# Create a DataFrame
df = pd.DataFrame(data, columns=columns)
```

```python
# Visualize the original data
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal length (cm)'], df.index)
plt.title("Original Sepal Length Data")
plt.xlabel("sepal length (cm)")
plt.ylabel("Index")
plt.show()
```

```python
# Calculate the mean and standard deviation
mean = df['sepal length (cm)'].mean()
std_dev = df['sepal length (cm)'].std()

# Define a threshold (e.g., 3 standard deviations)
threshold = 3 * std_dev

# Identify outliers
outliers = df[(df['sepal length (cm)'] - mean).abs() > threshold]

# Remove outliers
cleaned_df = df[(df['sepal length (cm)'] - mean).abs() <= threshold]

# Visualize the cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(cleaned_df['sepal length (cm)'], cleaned_df.index)
plt.title("Cleaned Sepal Length Data")
plt.xlabel("sepal length (cm)")
plt.ylabel("Index")
plt.show()
```
